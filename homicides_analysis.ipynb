## Countries with highest homicide rates

```python
import pandas as pd
sql_query = '''
SELECT 
    country, 
    homicides_2020, 
    rates_2020
FROM homicides 
WHERE rates_2020 IS NOT NULL 
ORDER BY rates_2020 DESC 
LIMIT 10;
'''

result = pd.read_sql_query(sql_query, conn)
display(result)
```

## Homicide Rate increase from 2019 to 2020

```python
import pandas as pd
sql_query = '''
SELECT 
    region,
    (rates_2020 - rates_2019) AS rate_increase
FROM homicides
WHERE rates_2020 IS NOT NULL AND rates_2019 IS NOT NULL
ORDER BY rate_increase DESC
LIMIT 10;
'''

result = pd.read_sql_query(sql_query, conn)
display(result)
```

```python
import pandas as pd
import numpy as np

# Assuming that 'conn' is a valid connection to the SQLite database
sql_query = '''SELECT * FROM homicides;'''

# read data from the database
homicides_df = pd.read_sql_query(sql_query, conn)

# list of columns for which to keep non-NaN values
cols_to_keep = ['Region', 'Subregion', 'country', 'Source']
cols_to_keep.extend([f'homicides_{i}' for i in range(2010, 2021)])
cols_to_keep.extend([f'rates_{i}' for i in range(2010, 2021)])

# create a new dataframe with non-NaN values for selected columns
homicides_clean_df = homicides_df[cols_to_keep].dropna()

# create a new dataframe with only NaN values for selected columns
cols_to_check = [f'homicides_{i}' for i in range(2010, 2021)]
cols_to_check.extend([f'rates_{i}' for i in range(2010, 2021)])
homicides_nans_df = homicides_df[homicides_df[cols_to_check].isna().any(axis=1)]

# print the first 5 rows of each dataframe
print('Cleaned dataframe:')
display(homicides_clean_df.head())
print('\nDataframe with NaN values:')
display(homicides_nans_df.head())
```
