## Countries with highest homicide rates

```python
import pandas as pd
sql_query = '''
SELECT 
    country, 
    homicides_2020, 
    rates_2020
FROM homicides 
WHERE rates_2020 IS NOT NULL 
ORDER BY rates_2020 DESC 
LIMIT 10;
'''

result = pd.read_sql_query(sql_query, conn)
display(result)
```

## Homicide Rate increase from 2019 to 2020

```python
import pandas as pd
sql_query = '''
SELECT 
    region,
    (rates_2020 - rates_2019) AS rate_increase
FROM homicides
WHERE rates_2020 IS NOT NULL AND rates_2019 IS NOT NULL
ORDER BY rate_increase DESC
LIMIT 10;
'''

result = pd.read_sql_query(sql_query, conn)
display(result)
```

## Cleaning the database

```python
import pandas as pd
import numpy as np

# Assuming that 'conn' is a valid connection to the SQLite database
sql_query = '''SELECT * FROM homicides;'''

# read data from the database
homicides_df = pd.read_sql_query(sql_query, conn)

# list of columns for which to keep non-NaN values
cols_to_keep = ['Region', 'Subregion', 'country', 'Source']
cols_to_keep.extend([f'homicides_{i}' for i in range(2010, 2021)])
cols_to_keep.extend([f'rates_{i}' for i in range(2010, 2021)])

# create a new dataframe with non-NaN values for selected columns
homicides_clean_df = homicides_df[cols_to_keep].dropna()

# create a new dataframe with only NaN values for selected columns
cols_to_check = [f'homicides_{i}' for i in range(2010, 2021)]
cols_to_check.extend([f'rates_{i}' for i in range(2010, 2021)])
homicides_nans_df = homicides_df[homicides_df[cols_to_check].isna().any(axis=1)]

# print the first 5 rows of each dataframe
print('Cleaned dataframe:')
display(homicides_clean_df.head())
print('\nDataframe with NaN values:')
display(homicides_nans_df.head())
```

## Creating table as db.

```python
#import sqlite3

#conn = sqlite3.connect('C:/Users/EFKA-PC/Desktop/github1/Global_homicides/homicides_fin.db')
c = conn.cursor()

# Execute SQL command to create table
c.execute('''CREATE TABLE homicides_clean AS SELECT Region, Subregion, country, homicides_2010, 
homicides_2011,homicides_2012, homicides_2013, homicides_2014, homicides_2015, homicides_2016, 
homicides_2017, homicides_2018, homicides_2019, homicides_2020, rates_2010, rates_2011, rates_2012, 
rates_2013,rates_2014, rates_2015, rates_2016, rates_2017, rates_2018, rates_2019, rates_2020 
FROM homicides WHERE Region = "Europe" AND 
        NOT homicides_2010 IS NULL AND NOT homicides_2011 IS NULL AND 
        NOT homicides_2012 IS NULL AND NOT homicides_2013 IS NULL AND 
        NOT homicides_2014 IS NULL AND NOT homicides_2015 IS NULL AND 
        NOT homicides_2016 IS NULL AND NOT homicides_2017 IS NULL AND 
        NOT homicides_2018 IS NULL AND NOT homicides_2019 IS NULL AND 
        NOT homicides_2020 IS NULL AND 
        NOT rates_2010 IS NULL AND NOT rates_2011 IS NULL AND 
        NOT rates_2012 IS NULL AND NOT rates_2013 IS NULL AND 
        NOT rates_2014 IS NULL AND NOT rates_2015 IS NULL AND 
        NOT rates_2016 IS NULL AND NOT rates_2017 IS NULL AND 
        NOT rates_2018 IS NULL AND NOT rates_2019 IS NULL AND 
        NOT rates_2020 IS NULL; ''')

# Commit the changes
conn.commit()

# Close the connection
#conn.close()
```
